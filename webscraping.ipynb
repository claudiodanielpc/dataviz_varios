```python
#Librerías necesarias
import pandas as pd
#Librerías para webscraping
import requests
from bs4 import BeautifulSoup
```

```python
#URL para webscraping
basica="https://www.farmalisto.com.mx/2087-cancer"

headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.1; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.149 Safari/537.36'}
```

```python
#Hacer un loop para generar la respuesta para n páginas
soup=[]
productos=[]
for i in range(1,20):
    url=basica+"?page="+str(i)
    response=requests.get(url,headers=headers)
    soup.append(BeautifulSoup(response.content, "html.parser"))
    productos.append(soup[i-1].find_all("script",{"class":"product_gtm_data"}))
```

```python
#Llevar los datos a una dataframe
df=pd.DataFrame()
datos=[]
for producto in productos:
    for i in range(len(producto)):
        datos.append(producto[i].text)
        df=df.append(pd.read_json(producto[i].text,typ="series"),ignore_index=True)

#Precios como variable numérica
df["price"]=df["price"].astype(float)
```
